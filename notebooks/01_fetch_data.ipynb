```python
# importing necessary libraries
import pandas as pd 
import numpy as np
from pathlib import Path
```

```python
# import the dataset and read rows
data=pd.read_csv("../data/raw/crypto_data.csv")
data.head()
```

```python
# check the data info
data.info()
```

```python
data["open_time"] = pd.to_datetime(data["open_time"], unit='ms')
data["close_time"] = pd.to_datetime(data["close_time"], unit='ms')
data["close"] = data["close"].astype(float)
data.head()
```

```python
# drop columns
data.drop('ignore', axis=1, inplace=True)
```

```python
data["close_time"] = pd.to_datetime(data["close_time"], unit='ms')
```

```python
# save the cleaned data to processed folder
data.to_csv("../data/processed/crypto_data_processed.csv", index=False)
```
